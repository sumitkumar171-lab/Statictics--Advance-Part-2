```python
'''
Q-1> What is hypothesis testing in statistics ?
Sol:
Hypothesis testing is a method in statistics used to make decisions or inferences about a population based on sample data. 
It helps you determine whether there is enough evidence to support a specific claim or hypothesis about a population parameter
(like a mean or proportion).
Keys:
1.Null Hypothesis (H‚ÇÄ)
2.Alternative Hypothesis (H‚ÇÅ or Ha)
3.Significance Level (Œ±)
4.Test Statistic
5.P-value
6.Reject or Fail to Reject H‚ÇÄ
'''
```

```python
'''
Q-2> What is the null hypothesis, and how does it differ from the alternative hypothesis ?
Sol:
The null hypothesis is a statement of no effect, no difference, or status quo. It's the hypothesis that a statistical test seeks 
to disprove or test against. On other hand The alternative hypothesis is what you‚Äôre trying to prove. It represents a new claim, 
effect, or difference that contradicts the null hypothesis.
Null Hypothesis (H‚ÇÄ) :
1.Assumes no change or effect
2.Tested to possibly be rejected
3.Assumed to be true unless evidence suggests otherwise
4.There is no difference in test scores between two groups.
Alternative Hypothesis (H‚ÇÅ) :
1.Claims there is a change or effect
2.Supported if H‚ÇÄ is rejected
3.Not assumed true unless data strongly supports it
4.There is a difference in test scores between two groups.
'''
```

```python
'''
Q-3> What is the significance level in hypothesis testing, and why is it important ?
Sol:
The significance level is the probability of making a Type I error, which means rejecting the null hypothesis when it is actually true.
Common values of Œ±:
1. 0.05 (5%) ‚Äì most commonly used
2. 0.01 (1%) ‚Äì stricter
3. 0.10 (10%) ‚Äì more lenient
It is important because of the following reasons:
1.Controls Error Rate: It helps you control how often you're willing to make a wrong conclusion (false positive).
2.Decision Rule: It gives you a clear rule ‚Äî if the p-value < Œ±, you reject H‚ÇÄ; otherwise, you fail to reject H‚ÇÄ.
3.Balances Risk: It lets researchers decide how cautious they want to be. A lower Œ± means you require stronger evidence to reject H‚ÇÄ.
'''
```

```python
'''
Q-4> What does a P-value represent in hypothesis testing ?
Sol:
The p-value is the probability of getting a result as extreme or more extreme than your observed data, assuming the null hypothesis is true.
Interpreting the P-value:
    P-value	                    Interpretation
Low (e.g., < 0.05)	    Unlikely result if H‚ÇÄ is true ‚Üí reject H‚ÇÄ
High (e.g., > 0.05)	    Likely result if H‚ÇÄ is true ‚Üí fail to reject H‚ÇÄ
'''
```

```python
'''
Q-5> How do you interpret the P-value in hypothesis testing ?
Sol:
p-value tells us the likelihood of observing your sample data (or something more extreme), assuming the null hypothesis (H‚ÇÄ) is true.

P-value	                  Interpretation	                                                                       Conclusion
‚â§ Œ± (e.g., 0.05)	    The result is statistically significant. There‚Äôs strong evidence against H‚ÇÄ.	            Reject H‚ÇÄ
> Œ±	                    The result is not statistically significant. There‚Äôs not enough evidence to reject H‚ÇÄ.	    Fail to reject H‚ÇÄ
'''
```

```python
'''
Q-6> What are Type 1 and Type 2 errors in hypothesis testing ?
Sol:
Type I Error (False Positive) : Rejecting the null hypothesis (H‚ÇÄ) when it is actually true
1.Probability of this error: Denoted by Œ± (the significance level), e.g., 0.05.
Ex: A medical test says a person has a disease (positive), but they‚Äôre actually healthy.
Type II Error (False Negative):Failing to reject the null hypothesis (H‚ÇÄ) when the alternative hypothesis (H‚ÇÅ) is actually true.
1.Probability of this error: Denoted by Œ≤.
2.Power of a test = 1 ‚àí Œ≤ (how good your test is at detecting real effects).
Ex: A medical test says a person is healthy (negative), but they actually have the disease.

Error Type	  What Happens	    Reality	Test      Conclusion	       Risk Controlled By
Type I	     False Positive	    H‚ÇÄ is true	      Reject H‚ÇÄ	          Œ± (significance level)
Type II	     False Negative	    H‚ÇÅ is true	      Fail to reject H‚ÇÄ	  Œ≤ (Power = 1 ‚àí Œ≤)

'''


```

```python
'''
Q-7> What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
Sol:
One-Tailed Test (a.k.a. one-sided) :
1.It is used when a AI engineer only interested in one direction of an effect (either greater than or less than).
Hypotheses Example:
1.H‚ÇÄ: Œº = 100
2. H‚ÇÅ: Œº > 100 (Right-tailed)
or
H‚ÇÅ: Œº < 100 (Left-tailed)
Critical region:
1.Entire Œ± (e.g., 0.05) is placed on one side of the distribution.

Two-Tailed Test (a.k.a. two-sided)
It is used when a AI engineer want to detect any difference, regardless of direction.
Hypotheses Example:
1.H‚ÇÄ: Œº = 100
2.H‚ÇÅ: Œº ‚â† 100
Critical region: 
1.The Œ± level is split between both tails of the distribution (e.g., 0.025 on each side if Œ± = 0.05).
'''
```

```python
'''
Q-8> What is the Z-test, and when is it used in hypothesis testing ?
Sol:
A Z-test is a hypothesis test that uses the Z-distribution (a standard normal distribution) to determine if sample data significantly 
differ from a known population value or between groups.
It‚Äôs called a Z-test because the test statistic follows a Z-distribution, where the mean is 0 and standard deviation is 1.
Use a Z-test when:
1.AI engineer know the population standard deviation (œÉ)
2.AI engineer's sample size is large (n ‚â• 30), or the population is normally distributed
3.AI engineer is comparing:
4.A sample mean to a known population mean
5.Two population means (for independent samples)
Ex:
Proportions (pÃÇ vs. p)
'''
```

```python
'''
Q-9> How do you calculate the Z-score, and what does it represent in hypothesis testing ?
Sol:
Z-score (or Z-statistic) tells you how many standard deviations your sample statistic is from the population mean, assuming the null 
hypothesis is true.
In hypothesis testing, it helps you decide whether your sample result is typical or unusual under the assumption that the null hypothesis is correct.
Calculate the Z-score:
'''

sigma = 30          
n = 72              
x_bar = 45         
mu = 26            
z = (x_bar - mu) / (sigma / (n ** 0.5)) # Formula used
print(f"Z-score: {z:.2f}")
'''
Z-score	             Interpretation
Close to 0	        Sample mean is close to population mean ‚Äî not unusual
High (+Z)	        Sample mean is much higher than population mean
Low (‚àíZ)	        Sample mean is much lower than population mean
'''
```

```python
'''
Q-19> What is the F-distribution, and when is it used in hypothesis testing ?
Sol:
The F-distribution is a continuous probability distribution that arises frequently in hypothesis testing, especially when 
comparing variances or model fits. It's widely used in ANOVA (Analysis of Variance) and regression analysis.
Formula:
'''
from IPython.display import display, Math
display(Math(r"F = \frac{(X_1 / d_1)}{(X_2 / d_2)}"))
'''
Used:
1.ANOVA (Analysis of Variance)
2.Comparing Two Variances
3.Regression Model Significance
'''
```

```python
'''
Q-20> What is an ANOVA test, and what are its assumptions ?
Sol:
ANOVA tests the hypothesis:
Null Hypothesis H(base)0 : All group means are equal.
Alternative Hypothesis H(base)A : At least one group mean is different
Formula :
'''
from IPython.display import display, Math
display(Math(r"F = \frac{\text{Between-group variance}}{\text{Within-group variance}} = \frac{MS_{\text{between}}}{MS_{\text{within}}}"))
'''
Types of ANOVA :
Type	                                                                     Used When
One-way ANOVA	                                         Comparing means across 1 factor (e.g. test scores across 3 schools)
Two-way ANOVA	                                         Comparing means across 2 factors (e.g. diet and gender effects on weight loss)
Repeated Measures	                                     When the same subjects are tested under multiple conditions
Assumptions:
1.Independence of observations
2.Normality
3.Homogeneity of variances (Homoscedasticity)
'''
```

```python
'''
Q-21> What are the different types of ANOVA tests ?
Sol:
There are different types, each suited for different experimental designs and data structures. The main types of ANOVA tests are :
1.One-Way ANOVA : Tests for differences in means across three or more groups, based on one independent variable.
2.Two-Way ANOVA : Tests the effect of two independent variables, and their interaction on a dependent variable.
3.Repeated Measures ANOVA : Used when the same subjects are measured multiple times under different conditions or over time.
4.Two-Way Repeated Measures ANOVA : Combination of Two-Way ANOVA and Repeated Measures
5.Mixed-Design ANOVA (Split-Plot ANOVA) :  Used when you have both repeated measures and between-subjects factors.
6.MANOVA (Multivariate ANOVA) : Extends ANOVA when you have multiple dependent variables.
'''
```

```python
'''
Q-22> What is the F-test, and how does it relate to hypothesis testing ?
Sol:
The F-test compares two variances or tests multiple parameters simultaneously. It is based on the F-distribution, which is 
the ratio of two scaled chi-square distributions.
Formula:
'''
from IPython.display import display, Math
display(Math(r"F = \frac{\text{Variance}_1 / d_1}{\text{Variance}_2 / d_2}"))
'''
where :d (base)1 and d (base)2 are their respective degrees of freedom.
The Relation to Hypothesis Testing
The F-test is used to test null hypotheses of the form:
1.H(base)0 : Two variances are equal (or more generally, a simpler model is sufficient).
2.H(base)A : Two variances are not equal (or the complex model fits better)
'''
```

```python
'''
Q-1> Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
Sol:
'''
import numpy as np
from scipy.stats import norm
def z_test(sample, population_mean, population_std, alpha=0.05):
    sample = np.array(sample)
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    
    # Z-score
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")    
    if p_value < alpha:
        print("Reject the null hypothesis: Significant difference.")
    else:
        print("Fail to reject the null hypothesis: No significant difference.")
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=15, size=50)  # Sample drawn with a mean near 102
population_mu = 100
population_sigma = 15
z_test(sample_data, population_mu, population_sigma)
```

```python
'''
Q-5> Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing ?
Sol:
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05, two_tailed=True):
    se = sigma / np.sqrt(n)
    z_alpha = norm.ppf(1 - alpha / 2) if two_tailed else norm.ppf(1 - alpha)
    crit_val_right = mu0 + z_alpha * se
    crit_val_left = mu0 - z_alpha * se if two_tailed else None
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)
    null_pdf = norm.pdf(x, mu0, se)
    alt_pdf = norm.pdf(x, mu1, se)
    plt.figure(figsize=(12, 6))
    plt.plot(x, null_pdf, label='H‚ÇÄ Distribution (Œº=%.0f)' % mu0, color='blue')
    plt.plot(x, alt_pdf, label='H‚ÇÅ Distribution (Œº=%.0f)' % mu1, color='green')
    plt.fill_between(x, null_pdf, where=(x >= crit_val_right), color='red', alpha=0.3, label='Type I Error (Œ±)')
    if two_tailed and crit_val_left:
        plt.fill_between(x, null_pdf, where=(x <= crit_val_left), color='red', alpha=0.3)
    if two_tailed:
        beta_region = (x > crit_val_left) & (x < crit_val_right)
    else:
        beta_region = x < crit_val_right
    plt.fill_between(x, alt_pdf, where=beta_region, color='orange', alpha=0.3, label='Type II Error (Œ≤)')
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.axvline(crit_val_right, color='black', linestyle='--', label='Critical Value(s)')
    if two_tailed and crit_val_left:
        plt.axvline(crit_val_left, color='black', linestyle='--')
    plt.legend()
    plt.grid(True)
    plt.show()
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05, two_tailed=True)
```

```python
'''
Q-10> Write a Python program to calculate the margin of error for a given confidence level using sample data ?
Sol:
'''
import numpy as np
from scipy.stats import t
def calculate_margin_of_error(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    se = std_dev / np.sqrt(n)  
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_crit * se
    print(f"Sample Mean = {mean:.2f}")
    print(f"Sample Size = {n}")
    print(f"Sample Std Dev = {std_dev:.2f}")
    print(f"{int(confidence*100)}% Margin of Error = ¬±{moe:.2f}")
    return moe
np.random.seed(10)
sample_data = np.random.normal(loc=75, scale=12, size=40)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
'''
Q-11>  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process ?
Sol:
'''
def bayes_theorem(prior, sensitivity, specificity):
    p_disease = prior
    p_no_disease = 1 - prior
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity
    numerator = p_pos_given_disease * p_disease
    denominator = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)
    posterior = numerator / denominator
    print("Bayesian Inference Results:")
    print(f"P(Disease) = {p_disease}")
    print(f"P(Positive | Disease) = {p_pos_given_disease}")
    print(f"P(Positive | No Disease) = {p_pos_given_no_disease}")
    print(f"=> P(Disease | Positive) = {posterior:.4f}")
    return posterior
bayes_theorem(prior=0.01, sensitivity=0.99, specificity=0.95)
```

```python
'''
Q-13>  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data ?
Sol:
'''
import pandas as pd
import numpy as np
def calculate_expected_frequencies(observed):
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()
    expected = np.outer(row_totals, col_totals) / grand_total
    expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)    
    return expected_df
observed_data = pd.DataFrame({'Yes': [40, 20],'No': [10, 30]},index=['Group A', 'Group B'])
print("Observed Frequencies:")
print(observed_data)
expected = calculate_expected_frequencies(observed_data)
print("\nExpected Frequencies (assuming independence):")
print(expected.round(2))
```

```python
'''
Q-16> Implement an F-test using Python to compare the variances of two random samples ?
Sol:
'''
import numpy as np
from scipy.stats import f
def perform_f_test(sample1, sample2, alpha=0.05):    
    var1 = np.var(sample1, ddof=1)  
    var2 = np.var(sample2, ddof=1)    
    if var1 > var2:
        F = var1 / var2
        dfn = len(sample1) - 1
        dfd = len(sample2) - 1
    else:
        F = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1    
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))
    print("F-statistic:", round(F, 4))
    print("Degrees of freedom:", dfn, "and", dfd)
    print("P-value:", round(p_value, 4))
    if p_value < alpha:
        print(f"Reject the null hypothesis: the variances are significantly different (Œ± = {alpha})")
    else:
        print(f"Fail to reject the null hypothesis: no significant difference in variances (Œ± = {alpha})")
np.random.seed(0)
sample1 = np.random.normal(23,34,80)  
sample2 = np.random.normal(65,70,98)  
perform_f_test(sample1,sample2)
```

```python
'''
Q-16> Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results ?
Sol:
'''
import numpy as np
from scipy.stats import f_oneway
def perform_anova(*groups):    
    F_statistic, p_value = f_oneway(*groups)
    print("ANOVA Results")
    print("-" * 30)
    print(f"F-statistic: {F_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    alpha = 0.05
    if p_value < alpha:
        print(f"\nResult: Reject the null hypothesis (p < {alpha})")
        print("Conclusion: At least one group mean is significantly different.")
    else:
        print(f"\nResult: Fail to reject the null hypothesis (p ‚â• {alpha})")
        print("Conclusion: No significant difference between group means.")
np.random.seed(42)
group1 = np.random.normal(loc=24,scale=7,size=24)
group2 = np.random.normal(loc=82,scale=7,size=24)
group3 = np.random.normal(loc=39,scale=7,size=24)
perform_anova(group1,group2,group3)
```

```python
'''
Q-18> Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA ?
Sol:
'''
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene
import pandas as pd
def check_anova_assumptions(*groups):
    group_labels = [f'Group {i+1}' for i in range(len(groups))]
    flat_data = np.concatenate(groups)
    labels = np.concatenate([[label]*len(group) for label, group in zip(group_labels, groups)])
    df = pd.DataFrame({'Value': flat_data, 'Group': labels})
    print("=== Checking Assumptions for ANOVA ===\n")    
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        print(f"Shapiro-Wilk Test - {group_labels[i]}: p-value = {p:.4f} {'(Normal)' if p > 0.05 else '(Not normal)'}")
    stat, p = levene(*groups)
    print(f"\nLevene‚Äôs Test for Equal Variance: p-value = {p:.4f} {'(Equal variances)' if p > 0.05 else '(Unequal variances)'}")    
    print("\nIndependence is assumed to be true based on study design.\n")    
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(x='Group', y='Value', data=df)
    plt.title("Boxplot of Groups")
    plt.subplot(1, 2, 2)
    sns.histplot(data=df, x='Value', hue='Group', kde=True, element='step')
    plt.title("Histogram with KDE")
    plt.tight_layout()
    plt.show()
np.random.seed(1)
group1 = np.random.normal(45,2,25)
group2 = np.random.normal(1,2,25)
group3 = np.random.normal(90,200,25)
check_anova_assumptions(group1,group2,group3)
```

```python
'''
Q-19> Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results ?
Sol:
'''
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
np.random.seed(0)
n = 10  
factor_A = ['Low','High']
factor_B = ['Type1','Type2']
data = []
for a in factor_A:
    for b in factor_B:
        mean = 50
        if a == 'High': mean += 5
        if b == 'Type2': mean += 5
        if a == 'High' and b == 'Type2': mean += 5  # interaction effect
        scores = np.random.normal(loc=mean, scale=5, size=n)
        for score in scores:
            data.append({'Factor_A': a, 'Factor_B': b, 'Score': score})
df = pd.DataFrame(data)
model = ols('Score ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("=== Two-Way ANOVA Results ===\n")
print(anova_table)
plt.figure(figsize=(10, 6))
sns.pointplot(data=df, x='Factor_A', y='Score', hue='Factor_B', dodge=True, markers=["o", "s"], capsize=.1)
plt.title("Interaction Plot: Factor A vs Factor B on Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
'''
Q-26>  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results ?
Sol:
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)   
data2 = np.random.normal(loc=52, scale=15, size=100)   
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = 2 * min(stats.f.cdf(f_stat, df1, df2),1 - stats.f.cdf(f_stat, df1, df2))
print(f"Variance 1: {var1:.2f}, Variance 2: {var2:.2f}")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
plt.figure(figsize=(10,5))
plt.hist(data1, bins=20, alpha=0.6, label='Data 1 (std ‚âà {:.2f})'.format(np.std(data1, ddof=1)))
plt.hist(data2, bins=20, alpha=0.6, label='Data 2 (std ‚âà {:.2f})'.format(np.std(data2, ddof=1)))
plt.title("Histogram of Two Datasets")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
